```python
import pandas as pd
df = pd.read_csv('companies.csv')
df.head()
```

```python
# plt.figure(figsize = (10,6))
# sns.countplot(y='state', data=df)
# plt.title('Company Count by State')
# plt.xticks(rotation = 90)
# plt.show()

df_sorted = df['state'].value_counts().sort_values(ascending = False)

plt.figure(figsize = (10,6))
sns.countplot(y='state', data=df, order = df_sorted.index, color = 'teal')
plt.title('Company Count by State')
plt.xticks(rotation = 90)
plt.show()
```

```python
df.columns
```

```python
sns.barplot(y='Industry', x='job_openings', data=df, estimator = 'sum', errorbar=None)
plt.xticks(rotation = 90)
plt.title('Job Openings by Industry')
plt.show()
```

```python
# Create new data frame
df_rate = df.groupby('founded')['growth_percentage'].mean().reset_index()

sns.lineplot(x='founded', y='growth_percentage', data=df, estimator='average', errorbar = None)
plt.title('Average Growth Percentage by Year Founded')
plt.show()
```

```python
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.ylabel('estimated revenues (in millions)')
plt.title('Employees and Estimated Revenue')
plt.show()
```

```python
sns.histplot(x='founded', data=df, bins = 20)
plt.title('Number of Companies Founded Each Year')
plt.xlabel('Year Founded')
plt.show()
```
