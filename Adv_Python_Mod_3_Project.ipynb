```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/home.html').text
```

```python
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup.find('table')
```

```python
soup.find('table').find_all('tr', class_='book')
```

```python
id = []
title = []
author = []
language = []
pages = []
topic = []
price = []


for book in soup.find('table').find_all('tr', class_='book'):
    id.append(book.find_all('td')[0].text)
    title.append(book.find_all('td')[1].text)
    author.append(book.find_all('td')[2].text)
    language.append(book.find_all('td')[3].text)
    pages.append(book.find_all('td')[4].text)
    topic.append(book.find_all('td')[5].text)
    price.append(book.find_all('td')[6].text)
```

```python
title
```

```python
pd.DataFrame({
    'id': id,
    'title': title,
    'author': author,
    'language': language,
    'pages': pages,
    'topic': topic,
    'price': price
})
```

```python
df = pd.DataFrame({
    'id': id,
    'title': title,
    'author': author,
    'language': language,
    'pages': pages,
    'topic': topic,
    'price': price
})
```

```python
df.info()
```

```python
df['pages'] = pd.to_numeric(df['pages'])
```

```python
df['price'] = df['price'].str.replace('$', '', regex=False).str.replace(',', '')
```

```python
df['price'] = pd.to_numeric(df['price'])
```

```python
df['topic'].value_counts()
```

```python
avg_price_by_topic = df.groupby('topic')['price'].mean()
print(avg_price_by_topic)
```

```python
avg_pages_by_topic = df.groupby('topic')['pages'].mean()
print(avg_pages_by_topic)
```
