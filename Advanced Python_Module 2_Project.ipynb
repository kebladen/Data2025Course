```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('housing.csv')
df.head()
```

```python
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
plt.show()
```

```python
Avg_PricePerYear_df = df.groupby('YearBuilt', as_index=False).agg({'Price': 'mean'})
Avg_PricePerYear_df
```

```python
plt.plot(Avg_PricePerYear_df['YearBuilt'], Avg_PricePerYear_df['Price'])
plt.show()
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=df)
plt.show()
```

```python
Avg_PricePerRegion_df = df.groupby('Regionname', as_index=False).agg({'Price': 'mean'})
Avg_PricePerRegion_df
```

```python
sns.barplot(x='Regionname', y='Price', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(x='Price', data=df)
plt.show()
```

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df)
plt.show()
```
